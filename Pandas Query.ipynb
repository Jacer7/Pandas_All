>>Returns a object containing counts of unique values

 <font size="5"> Which country has won the most men's gold medals in singles badminton over the years?   
Also sort the results alphabetically by the player's names.?</font> 

```python
gbm = oo[(oo.Medal == 'Gold') & (oo.Gender == 'Men') & (oo.Sport=='Badminton') & (oo.Event == 'singles')]
gbm.sort_values(by=['Athlete', 'Edition'])
```

```python
#checking wheather badminton is under sports series or not
jay = oo.Sport.value_counts()
jay.sort_values()
```

<font size = "5"> Which three countries have won the most medals in recent years (from 1984 to 2008)?</font-size>


```python
recent = oo[oo.Edition >= 1984]
recent.NOC.value_counts().head(3)
```

<font size = "5">Display the male gold medal winners for the 100m track and field sprint event over the years.   
    List the results starting with the most recent.   
    Show the Olympic city, edition, athlete, and the country they represent.
       
    
</font>

```python
gmh = oo[(oo.Medal =='Gold') & (oo.Gender =='Men') & (oo.Event == '100m') | (oo.Event == 'Field Sprint')] 
gmh.sort_values('Edition', ascending=False)[['City', 'Edition', 'Athlete', 'NOC']]
```
