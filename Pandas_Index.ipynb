## Pandas Indexing

- The index object is an immutable array   
- Indexing allows you to access a row or column using a label
** because of this property, pandas is special as in other typical language, we cannot access the data using label for an array.**

Indexing

```python
import pandas as pd
```

```python
oo = pd.read_csv('olympics.csv', skiprows=4);
```

```python
oo.head()
```

**Index**

```python
type(oo.index)
```

```python
oo.index[100]
```

**set_index()**

   DataFrame.set_index(keys,drop=True,append=False,inplace=False,verify_integrity=False)

- Set the DataFrame using one or more columns
- set_index(keys, inplace=True)


```python
oo.head()
```

```python
oo.reset_index(inplace=False)
```

```python
oo.set_index('Athlete', inplace=True)
```

```python
oo.head()
```

Set will set the desired index and reset will reset the previous dataFrame index

**reset_index()**

- opposite of the set index

**sort_index()**

- DataFrame.sort_index(axis=0, level = None, ascending = True, inplace =False, ...by=None)
- Sort objects by a label along the axis

```python
ath = pd.read_csv('olympics.csv', skiprows=4);
ath.head()
```

```python
ath.set_index('Athlete', inplace=True)
ath.head()
```

```python
ath.sort_index(inplace=True)
ath.head()
```

```python
ath.sort_index(inplace=True, ascending=False) #sorting by descending order
ath.head()
```

**loc[.....]: label based index**

```python
#first set the index to athlete
ath.loc['BOLT, Usain']
```

- loc vs traditional method for finding data

```python
oo[oo.Athlete == 'BOLT, Usain'] #traditional approach for data to generate using the dataframe
# this will not run but the syntax is correct. 
# we need to reset the index to run this syntax
```

**iloc : integer position based(from 0 to length -1 of the axis)**
- allows traditional python slicing

```python
df = pd.read_csv('olympics.csv', skiprows=4);
```

```python
df.head()
```

```python
df.iloc[1700]
```

```python
df.iloc[[1542, 2390, 6000, 15000]]
```

```python
df.iloc[1:4] #traditional python slicing
```
